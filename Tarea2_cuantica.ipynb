```julia
using LinearAlgebra
using Plots
```

```julia
# Definomos la base

V = [1 0; 0 1]

# Ahora, lo que es el estado ϕ
theta = range(0, stop = 2*pi, length = 50)
n = length(theta)
Bs = zeros(n)
Cs = zeros(n)

for ang in 1:n
    ketphi = (V[:,1] + exp(im * theta[ang])*V[:,2])/sqrt(2)
    braphi = transpose(conj(ketphi))
    #Definimos cada uno de los operadores
    B = V[:,1]*transpose(V[:,2]) + V[:,2]*transpose(V[:,1])
    C = V[:,1]*transpose(V[:,1]) - V[:,2]*transpose(V[:,2])
    Bs[ang] = braphi*B*ketphi
    Cs[ang] = braphi*C*ketphi
end
# Calculamos las multiplicaciones
```

```julia
plot(theta, Bs, title = "Comportamiento de B", 
    xlabel = "θ",
    ylabel = "B",
    label = "cos(θ)")
```

```julia
plot(theta, Cs, title = "Comportamiento de C", 
    xlabel = "θ",
    ylabel = "C",
    label = :none,
    ylimits = (-1,1)
)
```
